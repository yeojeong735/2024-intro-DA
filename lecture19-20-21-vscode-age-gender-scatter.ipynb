```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []


name = input('궁금한 동네를 입력하세요: ')
# 제주특별자치도

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            male.append(int(row[i].replace(',', '')))
        for i in range(106, 207) :
            female.append(int(row[i].replace(',', '')))
        break

f.close()

plt.rc('font', family ='Malgun Gothic')

plt.plot(male, label='male', color='b')
plt.plot(female, label='female', color='r')

plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import csv, math
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []
size = []

name = input('궁금한 동네를 입력하세요: ')
# 제주특별자치도

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            male.append(int(row[i].replace(',', '')))
            female.append(int(row[i + 103].replace(',', '')))        
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+103].replace(',', ''))))
            #<- (루트)남성 수 + 여성 수
        break

f.close()

plt.rc('font', family ='Malgun Gothic')
plt.figure(figsize=(10, 5))
plt.scatter(male, female, c=range(len(male)), alpha=0.7, cmap='jet', s=size)
plt.colorbar()
plt.plot(range(max(male)),range(max(male)), 'g')

plt.xlabel('Male')
plt.ylabel('Female')
plt.grid(True)
plt.show()
```

### 내장 함수 map()

```python
def addsome(value):
    return value + 10

lst = [10, 23, 4, 7]
result = map(addsome, lst)
list(result)
```

```python
def myreplace(s):
    return s.replace(',', '')

lst = ['34,000', '1,999', '2,786', '1,000,000']
result = map(myreplace, lst)
list(result)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

result = []
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')
        
    row[4:] = map(int, row[4:])
    result.append(row[10]) # 오전 7시 승차 데이터

f.close()

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 3))
plt.bar(range(len(result)), result)
plt.xlabel('역')
plt.ylabel('승차 인원수')
plt.grid(True)
plt.show()
```

```python
# 오전 7시 승차 데이터를 오름차순으로 정렬 후 막대그래프로 시각화하기
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

result = []
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')
    row[4:] = map(int, row[4:])
    result.append(row[10])

f.close()

result.sort()
# result.sort(reverse=True)

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 3))
plt.bar(range(len(result)), result)
plt.xlabel('역')
plt.ylabel('승차 인원수')
plt.grid(True)
plt.show()
```

```python
# 오전 7시 승차 데이터를 오름차순으로 정렬하고, 막대그래프로 시각화하기
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

result = []
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))

f.close()

result.sort() # 오름차순 정렬
# sort(reverse=True) : 내림차순 정렬

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 3))
plt.bar(range(len(result)), result)
plt.xlabel('역')
plt.ylabel('승차 인원수')
plt.grid(True)
plt.show()
```

```python
# 오전 7~9시 승차 인원이 최대인 역 찾기
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

max_people = 0
max_station = ''
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')

    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > max_people:
        max_people = sum(row[10:15:2])
        max_station = row[3] + '(' + row[1] + ')'

f.close()


print(max_station, ':', max_people, '명')
```
